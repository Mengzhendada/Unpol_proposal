```python
def Get_Q2(Ebeam,x,y):
    m_p=0.938272
    S=m_p**2+2*m_p*Ebeam
    Q2=S*x*y
    return Q2
```

```python
x=0.31
y=0.6576
Eb=12
Q2=Get_Q2(Eb,x,y)
print(Q2)
```

```python
def Get_y(Ebeam,x,Q2):
    m_p=0.938272
    S=m_p**2+2*m_p*Ebeam
    y=Q2/x/S
    return y
```

```python
x=0.125
Eb=12
Q2=1.5
y=Get_y(Eb,x,Q2)
print(y)
```
