```python
def Get_Eprime(Ebeam,theta):
    M_e = 0.93827
    theta_r = np.pi*theta/180
    E_prime = Ebeam/(1+2*Ebeam/M_e*np.sin(theta_r/2)**2)
    return E_prime
```

```python
import ROOT
from ROOT import TCanvas, TFile, TProfile, TNtuple, TH1F, TH2F
Acceptance_electron = ROOT.TFile.Open("Acceptance/acceptance_solid_SIDIS_He3_electron_201701_1e7_output_final.root","READ")
h_theta_momentum_forward = Acceptance_electron.Get("acceptance_ThetaP_forwardangle")
h_theta_momentum_large = Acceptance_electron.Get("acceptance_ThetaP_largeangle")
c1 = TCanvas( 'c1', 'Acceptance' )
h_theta_momentum_forward.Draw("colz")
h_theta_momentum_large.Draw("colz same")
c1.Modified()
c1.Update()
from ROOT import gROOT
gROOT.GetListOfCanvases().Draw()
c1.SaveAs("results/acceptance.pdf")
```

```python
def Get_Acceptance_e(theta,momentum):
    acc = 0
    Acceptance_electron = ROOT.TFile.Open("Acceptance/acceptance_solid_SIDIS_He3_electron_201701_1e7_output_final.root","READ")
    h_theta_momentum_forward = Acceptance_electron.Get("acceptance_ThetaP_forwardangle")
    h_theta_momentum_large = Acceptance_electron.Get("acceptance_ThetaP_largeangle")
    acc+=h_theta_momentum_forward.GetBinContent(h_theta_momentum_forward.GetXaxis().FindBin(theta),h_theta_momentum_forward.GetYaxis().FindBin(momentum))
    acc+=h_theta_momentum_large.GetBinContent(h_theta_momentum_large.GetXaxis().FindBin(theta),h_theta_momentum_large.GetYaxis().FindBin(momentum))
    return acc
```

```python
def Check_Acceptance_e(theta,momentum,Nbins):
    acc = 0
    Acceptance_electron = ROOT.TFile.Open("Acceptance/acceptance_solid_SIDIS_He3_electron_201701_1e7_output_final.root","READ")
    h_theta_momentum_forward = Acceptance_electron.Get("acceptance_ThetaP_forwardangle")
    h_theta_momentum_large = Acceptance_electron.Get("acceptance_ThetaP_largeangle")
    h_theta_momentum_forward.Rebin(Nbins)
    h_theta_momentum_large.Rebin(Nbins)
    acc+=h_theta_momentum_forward.GetBinContent(h_theta_momentum_forward.GetXaxis().FindBin(theta),h_theta_momentum_forward.GetYaxis().FindBin(momentum))
    acc+=h_theta_momentum_large.GetBinContent(h_theta_momentum_large.GetXaxis().FindBin(theta),h_theta_momentum_large.GetYaxis().FindBin(momentum))
    return acc/Nbins

```
